## Exploratory Data Analysis

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
salesByGender = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)
sns.barplot(x= 'Gender', y= 'Amount', data = salesByGender)
```

#### From the above both Graph we got to know that most of the buyers are females and even the purchasing power of femaleis higher than male

```python
#On the basis of Age
```

```python
ax = sns.countplot(x = 'Age Group', data = df, hue= 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount by Age Group
ax = sns.countplot(x = 'Age Group', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
salesByAge = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)
sns.barplot(x= 'Age Group', y= 'Amount', data = salesByAge)
```

#### From the Above graph we got to know that most of the byers are of age group between 26-35 year and mostly female

```python
#On the basis of State
```

```python
salesByState = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = salesByState, x= 'State', y= 'Orders')
```

```python
salesByState = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = salesByState, x= 'State', y= 'Amount')
```

```python
#Occupation
```

```python
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.countplot(data = df, x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
salesByState = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = salesByState, x= 'Occupation', y= 'Amount')
```

#### From the above graph we got to know that must of the buyers from IT Sector, Healthcare and Aviation

```python
#On the basis of product category
```

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df, x = 'Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
salesByState = df.groupby(['Category'],as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = salesByState, x= 'Category', y= 'Amount')
```

#### From the above graph we got to know that must of the Amount expend on Food 

```python
salesByState = df.groupby(['Product_ID'],as_index = False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = salesByState, x= 'Product_ID', y= 'Orders')
```

### Married female from age group 26-35 from UP, maharastra who is working in IT Sector, Healthcare are more likely to buy product from food and clothing category
